```python
import json
import os
```

```python
def correcting_response_files():
    parent_dir = "openpain.org/subacute_longitudinal_study/"
    for file in sorted(os.listdir(parent_dir)):
        if "task" in file:
            file_path = os.path.join(parent_dir, file)
            with open(file_path, "r") as f:
                data = json.load(f)
            try:
                slice_timing = data["SliceTiming"]
                new_slice_timing = [x / 1000 for x in slice_timing]
                data["SliceTiming"] = new_slice_timing
                
                print(data["SliceTiming"])
                
                for element in data["SliceTiming"]:
                    if element > data["RepetitionTime"]:
                        print("Element found greater than repitition", element)
                
                with open(file_path, "w") as f:
                    json.dump(data, f)
            except:
                print("slicing time not available")
```

```python
correcting_response_files()
```
